```python
# Write a program to generate and print 10 random numbers between 20 and 50 find and
#print all the even numbers from these randomly generated numbers

import random
rand_list = []
even_list = []

for i in range (10):
    number = random.randint(20,50)
    rand_list.append(number)

print(rand_list)

#queremos encontrar los even numbers from these randomly generated numbers

for i in rand_list:
    if i%2==0:
        even_list.append(i)

print(even_list)
    
```

```python
#Write a program to find and print all the prime numbers between a given range, for example between 20 and 50

lower= 20
upper= 50

print("Prime numbers between lower and upper are:")

#flag es una variable
#se pone 2 porque los numeros primos empiezan de 2 en adelante
#n es la variable entre lower y upper

for n in range(lower, upper+1):
    flag = 0
    for i in range(2,n):
        if n%i==0:
            flag = 1
    if flag==0:
        print(n)
    
```
