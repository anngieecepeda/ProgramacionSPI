```python
# Task 4
# Import the required modules

import pandas as pd
import matplotlib.pyplot as plt

def load_data(file):
    df= pd.read_csv('dataset-for-Task4.csv')
    return df

def data_description(df):
    df.head()
    print("Summary Dataframe:")
    print(df.describe())

def bargraph(df):
    average = df.groupby("Country")["Sales"].mean()
    average.plot.bar()
    plt.title("Average Sales per Country")
    plt.xlabel("Country")
    plt.ylabel("Average Sales")
    plt.show()

def piechart(df):
    g1_0_20 = df[df["Age"] <= 20]["Sales"].sum()
    g2_21_30 = df[(df["Age"] > 20) & (df["Age"] <= 30)]["Sales"].sum()
    g3_31_40 = df[(df["Age"] > 30) & (df["Age"] <= 40)]["Sales"].sum()
    g4_41_50 = df[(df["Age"] > 40) & (df["Age"] <= 50)]["Sales"].sum()
    g5_51_more = df[df["Age"] > 50]["Sales"].sum()

    #List
    Sales = [g1_0_20, g2_21_30, g3_31_40, g4_41_50, g5_51_more]
    Groups = ['0-20', '21-30', '31-40', '41-50', '51+']

    plt.pie(Sales, labels=Groups)
    plt.title("Distribution of Different Age Groups ")
    plt.show()

def histogram(df):
    g1_0_20 = len(df[df["Age"] <= 20])
    g2_21_30 = len(df[(df["Age"] > 20) & (df["Age"] <= 30)])
    g3_31_40 = len(df[(df["Age"] > 30) & (df["Age"] <= 40)])
    g4_41_50 = len(df[(df["Age"] > 40) & (df["Age"] <= 50)])
    g5_51_more = len(df[df["Age"] > 50])

    #List2
    Counts = [g1_0_20, g2_21_30, g3_31_40, g4_41_50, g5_51_more]
    Groups = ['0-20', '21-30', '31-40', '41-50', '51+']
    
    plt.bar(Groups, Counts, color='salmon')
    plt.title("Distribution of Different Age Groups")
    plt.xlabel("Age Groups")
    plt.ylabel("Number of People")
    plt.show()
    
    
# Main Program

def main():
    file = 'dataset-for-Task4.csv'
    df = load_data(file)
    data_description(df)
    bargraph(df)
    piechart(df)
    histogram(df)

main()
```

```python
#Task 5

    
# Code for creating Classes

class Vehicle:
    def __init__ (self, make, model, year, color):
        self.make= make
        self.model= model
        self.year= year
        self.color= color
    
    def display_details(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Color: {self.color}")
        

class Car(Vehicle):
    def __init__(self, make, model, year, color, doors, kind):
        super().__init__(make, model, year, color)
        self.doors= doors
        self.kind= kind
    def display_details(self):
        super().display_details()
        print(f"Doors: {self.doors}")
        print(f"Type: {self.kind}")

class Truck(Vehicle):
    def __init__(self, make, model, year, color, payload_capacity):
        super().__init__(make, model, year, color)
        self.payload_capacity= payload_capacity 
    def display_details(self):
        super().display_details()
        print(f"Payload Capacity: {self.payload_capacity} kg")

class Motorcycle(Vehicle):
    def __init__(self, make, model, year, color, engine_displacement, kind):
        super().__init__(make, model, year, color)
        self.engine_displacement= engine_displacement
        self.kind= kind
    def display_details(self):
        super().display_details()
        print(f"Engine Displacement: {self.engine_displacement} CC")
        print(f"Kind: {self.kind}")


# Code for creating objects

car1 = Car("Honda", "Civic", 2019, "Green", 5, "Hatchback")
truck1 = Truck("Dodge", "RAM 1500", 2021, "Silver", 3500)
moto1= Motorcycle("Ducati", "Panigale", 2018, "Red", 1103, "Sportbike")

#Call 

print("Car Details:")
car1.display_details()

print("\nTruck Details:")
truck1.display_details()

print("\nMotorcycle Details:")
moto1.display_details()


```
